```python
import pandas as pd
```

```python
nlo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')
# nlo_data
```

```python
nlo_df = nlo_data.copy()
nlo_df
```

```python
nlo_df['Time'] = pd.to_datetime(nlo_df['Time'])   # приводим к стандартному виду datetime
nlo_df
```

```python
nlo_df['Time'].dt.year.value_counts()    # в каком году больше всего было замечено НЛО
```

```python
import numpy as np
nlo_df_nevada = nlo_df[nlo_df['State']=='NV'].copy()                          # создаём базу по штату Невада
nlo_df_nevada['Date'] = pd.to_datetime(nlo_df_nevada['Time'].dt.date)         # создаём столбец с датами
nlo_df_nevada['Interval'] = nlo_df_nevada['Date'].diff().dt.days     # Интервал в днях между соседними событиями
# display(nlo_df_nevada)
interval = nlo_df_nevada['Interval'][1:].astype('int64')
round(interval.mean())
```
